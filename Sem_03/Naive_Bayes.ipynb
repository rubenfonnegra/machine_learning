````
import numpy as np
import matplotlib.pyplot as plt
from sklearn.naive_bayes import GaussianNB
from sklearn.datasets import load_iris # Iris dataset
````

````
iris = load_iris() # Load the iris dataset
iris_x = iris.data[:,0:2]  # Features. We'll pick 2
iris_y = iris.target # Labels

# Print shapes
print ( iris_x.shape,iris_y.shape )
````

````
#Plot Dispersion
_,ax = plt.subplots(figsize=(5,4))
ax.scatter (iris_x[:,0],iris_x[:,1], c=iris_y,edgecolors='k', cmap='Paired')
````

````
# Instance kNN object
naive_bayes = GaussianNB()

# Training
naive_bayes.fit( iris_x , iris_y )
````

````
### Plot decision boundaries
def Boundaries(NamePred,horizontal, vertical,NValue, Nlabel, width, long):
  h_min,h_max = horizontal[:,0].min()-0.1, horizontal[:,0].max()+0.1
  v_min,v_max = vertical[:,1].min()-0.1, vertical[:,1].max()+0.1

  #cuadricula
  h_grid, v_grid = np.meshgrid(np.linspace(h_min,h_max,NValue), np.linspace(v_min,v_max,NValue))

  #calcular predicciones
  pred_grid = NamePred.predict(np.c_[h_grid.ravel(),v_grid.ravel()])

  #reshape predicion
  pred_grid = pred_grid.reshape(h_grid.shape)
  print(h_grid.shape, v_grid.shape, pred_grid.shape)

  #Plot Dispersion
  _, ax =plt.subplots(figsize = (width,long))
  ax.pcolormesh(h_grid, v_grid, pred_grid, cmap='Paired')
  ax.scatter(horizontal[:,0],vertical[:,1], c=Nlabel, edgecolors='k', cmap='Paired')
  ax.set_xticks([])
  ax.set_yticks([])
````

````
#Plot decision boundaries
Boundaries(naive_bayes,iris_x,iris_x,200, iris_y,5,4)
````
